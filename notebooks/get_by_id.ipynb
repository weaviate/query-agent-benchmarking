```python
import os
import weaviate

weaviate_client = weaviate.connect_to_weaviate_cloud(
    cluster_url=os.getenv("WEAVIATE_URL"),
    auth_credentials=weaviate.auth.AuthApiKey(os.getenv("WEAVIATE_API_KEY"))
)

weaviate_client.is_ready()
```

```python
my_collection = weaviate_client.collections.get("BrightBiology")
```

```python
from weaviate.classes.query import Filter
import time

start = time.time()
results = my_collection.query.fetch_objects(
    filters=Filter.by_property("dataset_id").equal(
        "eye_capture_shape/how-the-eye-capture-images_17_0.txt"
    ),
    limit=1
)
print(f"Retrieved in {time.time() - start} seconds")
print(len(results.objects))
```
