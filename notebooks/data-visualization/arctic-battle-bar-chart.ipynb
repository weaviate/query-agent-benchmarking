```python
import numpy as np
import matplotlib.pyplot as plt

# ---- Data ----
categories = [
    "Biology", "Earth Science", "Economics", "Psychology", "Robotics"
]
arctic_2 =  [0.1451, 0.242, 0.182, 0.211, 0.156]
arctic_1_5 =  [0.1388, 0.163, 0.106, 0.108, 0.139]

x = np.arange(len(categories))
width = 0.35

# ---- Styling defaults ----
plt.rcParams.update({"font.size": 12, "font.family": "DejaVu Sans"})

fig, ax = plt.subplots(figsize=(10, 6))

# Bars
bars1 = ax.bar(
    x - width/2,
    arctic_2,
    width,
    label="Arctic 2.0 (+Hybrid Search)",
    color="#4B9CD3"  # Carolina Blue
)
bars2 = ax.bar(
    x + width/2,
    arctic_1_5,
    width,
    label="Arctic 1.5 (+Hybrid Search)",
    color="#001A57"  # Navy Blue
)

centers2 = [b.get_x() + b.get_width()/2 for b in bars2]

# Percent labels on bars
def add_bar_labels(bars, pad=0.01):
    for b in bars:
        h = b.get_height()
        ax.text(
            b.get_x() + b.get_width()/2,
            h + pad,
            h,
            ha="center", va="bottom"
        )

add_bar_labels(bars1)
add_bar_labels(bars2)

# Title, axes, legend
ax.set_title("Recall @ 5 Comparison Across Benchmarks", fontsize=14, pad=15)
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=15, ha="right")
ax.set_ylim(0, 1.02)  # small headroom for labels
ax.legend()

# Clean look: grid + spines
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
#ax.yaxis.grid(True, linestyle="--", alpha=0.7)

plt.tight_layout()
plt.savefig("arctic-battle-recall-at-5-comparison.png")
plt.show()
```
